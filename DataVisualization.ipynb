```python
import matplotlib.pyplot as plt
import seaborn as sns

#load titanic data set
titanic_df = sns.load_dataset('titanic')
#count the number of female and male in the dataset
gender_count = titanic_df['sex'].value_counts()
gender_count.plot(kind='bar', title="Sex Distribution")
plt.xlabel('sex')
plt.ylabel('counts')
plt.show()
#set the seaborn default aesthetic paramters



```

```python
class_data = titanic_df['pclass'].value_counts() #passenger class distribution
class_data.plot(kind='bar', title="Passenger class distribution")
plt.xlabel("Passenger Class")
plt.ylabel("count")
plt.show()
```

```python
#Embarked point distribution
embarked_data = titanic_df['embarked'].value_counts()
#color sets the color of the plot
#alpha sets the transparency level
#whether or not to display grid lines
embarked_data.plot(kind='bar', color='skyblue', alpha=0.7, grid=True)
plt.xlabel('Embarkation point')
plt.ylabel("count")
plt.title("Embarkation Point Distribution")
plt.show()
```

five preset Seaborn themes: darkgrid, whitegrid, dark, white, ticks

```python
#set the seaborn default aesthetic paramters
sns.set(style='whitegrid', palette='Blues',font='serif',font_scale=1.2)
#create a plot
plt.figure(figsize=(12,6))
sns.countplot(x='pclass',data=titanic_df, hue='pclass',palette='coolwarm', legend=False)
plt.title("Passenger Class Count")
plt.xlabel("passenger Class")
plt.ylabel('Count')
plt.legend(title='passenger class')
plt.xticks(rotation=45) #shorten the labels
plt.show()
```

Customizing The Histogram

```python
#increase the number of bins to 30
sns.histplot(data=titanic_df, x='age', bins=30, kde=True)
#add a title
plt.title("Age Distribution Among Titanic Passengers")
#label the x and y axis
plt.xlabel('Age')
plt.ylabel("Number of passengers")
```

Using hue to color the histogram bars for the age depending on gender
Multiple can be set to 'stack', 'fill', "dodge", "layer"
"stack" : Draw one histogram, stack the values of each variation on top of the other
FILL : Draw one histogram, with area of each filled up to the total height
DODGE: Draw on histogram but dodge them
Palette : allows to chamge colors
binWidth : allows to set the width of the bins
Element helps change the appearance, you can set the element to "step" and "poly"

```python
#a histogram using hue, palette,and multiple
sns.histplot(data=titanic_df, x='age', hue='sex', multiple='stack',palette='pastel')
plt.title("Age Distribution Among Titanic Passenges")
```

```python
#A histgram with binwidth, and element
sns.histplot(data=titanic_df, x='age', binwidth=1, element='step', color='purple')
plt.title("Age Distribution Among Titanic Passengers")
#label the x and y axis
plt.xlabel('Age')
plt.ylabel("Number of passengers")
```

Bar Plot CUstomizations With Seaborn
hue  Represents an additional categorical variable by colors
color: Sets a specific color for all the plot bars
Order and hue_order : Arrange bars in a specific order
Orient: changes the plot's orientation


```python
#color-coded bar plot representing sex and survival ('survived')
sns.countplot(data=titanic_df, x='sex', color='cyan', order=['female', 'male'],
              orient='v').set_title('Sex and Survival Rates')
```

```python
#comparing the sex variable wuth survived
sns.countplot(x='sex', hue='survived', data=titanic_df)
```

```python
#comparing the pclass variable with survived
sns.countplot(x='pclass', hue='survived', data=titanic_df)
```

```python
#comparing embarkation point with survived
sns.countplot(x='embarked', hue='survived', data=titanic_df)
```

Multivariate Analysis Arena. Scatter Plots and correlation of Variables

```python
sns.scatterplot(x='age', y='fare', data=titanic_df)
plt.title('Age vs Fare')
plt.show()
```

```python
#scatter plot with age, fare and passenger class
sns.scatterplot(x='age', y='fare', hue='pclass', data= titanic_df)
plt.title("Age vs fare (separate colors for passenger class")
plt.show()
```

```python
#adding markers and sizes to the scatter plot
#sizes:  set the range of sizes to scale the scatter plot
#style: Makes different marks on the plot for different categories
sns.scatterplot(x='age',y='fare', hue='pclass',style='sex', size='fare',sizes=(20,200), data=titanic_df)
plt.title('Age vs Fare (Separate Markers for Sex and Sizes for Fare')
plt.show()
#correlation of all numerical variables in the dataset
corr_vals = titanic_df.corr(numeric_only=True)
print(corr_vals)
```

DIVING INTO BOX PLOTS

```python
sns.boxplot(x='pclass', y='fare', hue='survived',data=titanic_df)
plt.title("Fares Vs passenger class differentiated by survival")
plt.show()
```

```python
#tweaking the boxplot
sns.boxplot(x='survived', y='fare', hue='pclass', data=titanic_df, palette='Set3',
            linewidth=1.5)
plt.title("Survival and Passenger classes by fare")
plt.show()
```

```python
#customizing the boxplot more
sns.boxplot(x='pclass', y='fare', hue='survived',data=titanic_df, palette="Set3",
            linewidth=1.5, hue_order=[1,0], order=[3,1,2], color='skyeblue',
            saturation=0.7,dodge=True, fliersize=5)
plt.title('Fare vs passenger class differentiated by survival')
plt.show()
```

Inntroduction To correlation in pythpn

```python
correlation_matrix =titanic_df.corr(numeric_only = True)
print(correlation_matrix)
```

Correlation coefficients in the matrix depict the relationship between variables, and they lie in the range of -1 to 1 range.
When two features have a positive high correlation, their values fail tend to raise and fall together
On the other hand, a negative correlation, one feature's value raises, the other one tends to fail. correlation of 0 or closer means there's no linear relationship

HeatMAPS in Seaborn.
HeatMaps plot displays the numerical tabular data where cells are colored depending on the contained value

```python
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
#add cbar to the heat map to draw a colorbar
#vmin, vmax Establish the colormap limits
sns.heatmap(correlation_matrix, annot=True, cbar=True, vmin=-1, vmax=1)
plt.show()
```

```python
#using the cmpa to define colormap for the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
#colorwarm contrast between negative and positive extremes, colormap scale goes from -1 to 1 in corresponding with correlation coefficient range
```

```python
#build acolor map
color_map = sns.diverging_palette(220,20, as_cmap=True)
sns.heatmap(correlation_matrix, annot=True, cmap=color_map)
plt.show()
#The argument 220 and 20 denote the hues in degrees on the color wheel, starting from 0 to 360.
#220 refers to blue gue, 20 refers to orange
```
